```
# 4. Start Flask + ngrok tunnel
import os, threading
os.chdir('/content/DIP project')

from pyngrok import ngrok

# Optional: set your ngrok authtoken for stable tunnels
# ngrok.set_auth_token('YOUR_TOKEN_HERE')

public_url = ngrok.connect(5000)
print(f'\n✦  Open this URL in your browser:\n   {public_url}\n')

# Run Flask in a thread so the cell stays responsive
def run_flask():
    os.system('python app.py')

t = threading.Thread(target=run_flask, daemon=True)
t.start()

import time; time.sleep(3)
print('Server started. Use the URL above.')
```
